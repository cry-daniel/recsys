```python
sorted_vf_cnt = vf_cnt.sort_index()
cdf = sorted_vf_cnt.cumsum() / sorted_vf_cnt.sum()
```

```python
cdf
```

```python
import numpy as np
from matplotlib import pyplot as plt
cdf_bucket_means = np.array([b.max() for b in np.array_split(cdf.values, 200)])
plt.plot(range(200), cdf_bucket_means)
plt.xlabel("Video counts")
plt.ylabel("CDF")
plt.title("CDF of Video counts")
plt.grid(True)
plt.show()
```
